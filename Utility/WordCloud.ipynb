```python
# Python3 program to implement
# the above approach
# Import the following modules

# pip install matplotlib
import matplotlib.pyplot as plt
import random
# pip install wordcloud
from wordcloud import WordCloud, STOPWORDS
import numpy as np
from PIL import Image

def grey_color_func(word, font_size, position, orientation, random_state=None, **kwargs):
    return "hsl(0, 0%%, %d%%)" % random.randint(60, 100)

# Give the whole path of the text file,
# open it, read it, and encode it.
text = open(r'C:\Users\PH\Desktop\005.txt', mode = 'r', encoding = 'utf-8').read()

# The Image shape in which you wanna convert it to.
mask = np.array(Image.open(r'C:\Users\PH\Desktop\eyewhitebg.png'))

path = r"C:\Users\PH\Desktop\VL-Gothic-Regular.ttf"

# Now inside the WordCloud, provide some functions:
# stopwords - For stopping the unuseful words
# like [,?/\"]
# font_path - provide the font path to which you
# wanna convert it to.
# max_words - Maximum number of words in the
# output image. Also provide height and width
# of the mask

wc = WordCloud(stopwords = STOPWORDS,
			mask = mask,
            font_path = path,
			background_color = "black",
			max_words = 9000,
			max_font_size = 150,
			random_state = 100,
			width = mask.shape[1],
			height = mask.shape[0])

# Finally generate the wordcloud of the given text
plt.figure( figsize=(20,10), facecolor='k')
wc.generate(text)
plt.imshow(wc.recolor(color_func=grey_color_func, random_state=3), interpolation = "None")

# Off the x and y axis
plt.axis('off')


plt.tight_layout(pad=0)

# Now show the output cloud
plt.show()
```
